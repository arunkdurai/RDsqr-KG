Extract pathway gene from SMPDB

```python
import os
import pandas as pd
```

```python
all_files = [file for file in os.listdir(r"...path_to\RDsqr-KG\SMPDB\smpdb_proteins.csv") if file.endswith('.csv')]
```

```python
pathway_gene = []
```

```python
for file in all_files:
    file_path = os.path.join(r"...path_to\smpdb_proteins.csv", file)
    df = pd.read_csv(file_path)
    pathway_gene.append(df)
```

```python
pathway_gene
```

```python
df = pd.concat(pathway_gene, ignore_index=True)
```

```python
df.shape
```

```python
df.to_csv(r"...path_to\RDsqr-KG\Preprocessed_datasets\pathway_gene.csv", index=False)
```
